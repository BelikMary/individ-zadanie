### Понаблюдаем и сделаем вывод, что всего у нас представлено 298 аэропортов. Найдем топ-100 аэропортов с самым высоким пассажирским трафиком.

```python
traffic.head()
```

```python
traffic['True whole year']=traffic['January'] + traffic['February'] + traffic['March'] + traffic['April'] + traffic['May']+ traffic['June'] + traffic['July'] + traffic['August'] + traffic['September'] + traffic['October'] + traffic['November'] + traffic['December']
```

```python
traffic1=traffic.groupby(['Airport name'])['True whole year']. sum().reset_index()
traffic1=traffic1.sort_values(by='True whole year', ascending=False).head(100)
traffic1
```

```python
df=traffic
o=traffic1['Airport name']
df=df.loc[df['Airport name']. isin(o)]
df
```

```python
df2 = df.melt(id_vars = ['Airport name', 'Year'],
         value_vars = ['January', 'February', 'March', 'April', 'May','June', 'July', 'August', 'September', 'October', 'November','December'])

df2.rename(columns = {'variable':'Month', 'value':'Traffic'}, inplace = True)
```

```python
Months = ['January', 'February', 'March', 'April', 'May','June', 'July', 'August', 'September', 'October', 'November','December']
df2['Month'] = pd.Categorical(df2['Month'], categories=Months, ordered=True)
df2
```

Сводная таблица позволяет отследить тенденцию изменения пассажирских поток в отдельных аэропортах (визуализация представлена на Дашборде)

```python
port_year = traffic.pivot_table(index='Airport name', columns='Year', values='True whole year')

port_year=port_year.join(traffic1.set_index('Airport name'), on='Airport name').sort_values(by='True whole year', ascending=False).head(12)
port_year
```
